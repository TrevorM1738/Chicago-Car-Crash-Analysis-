```python
# Import necessary functions
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cleaned_data2.csv')
df.head()
df.columns
```

```python
df.drop(['Unnamed: 0','PERSON_TYPE', 'SEX','DEVICE_CONDITION','TRAFFICWAY_TYPE', 'ALIGNMENT', 'LATITUDE', 'LONGITUDE', 'LOCATION','BEAT_OF_OCCURRENCE', 'MOST_SEVERE_INJURY',
       'INJURIES_TOTAL', 'INJURIES_FATAL', 'INJURIES_INCAPACITATING',
       'INJURIES_NON_INCAPACITATING', 'INJURIES_REPORTED_NOT_EVIDENT',
       'INJURIES_NO_INDICATION', 'INJURIES_UNKNOWN','DAMAGE','SAFETY_EQUIPMENT',
       'AIRBAG_DEPLOYED', 'EJECTION', 'INJURY_CLASSIFICATION','SEC_CONTRIBUTORY_CAUSE'], axis=1, inplace=True)
df.head(20)
```

```python
df1 = pd.get_dummies(df)
df1.head()
```

```python
df1.drop(['INJURIES_INCAPACITATING INJURY','INJURIES_MINOR INJURY','INJURIES_NO INJURY'], axis=1, inplace=True)
```

```python
y = df1['INJURIES_FATAL']
X = df1.drop(columns=['INJURIES_FATAL'], axis=1)
```

```python
X = X.apply(lambda x : (x - x.min()) /(x.max() - x.min()), axis=0)
X.head()
```

```python

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# Instantiate the model
logreg = LogisticRegression(fit_intercept=False, C=1e12, solver='liblinear')

# Fit the model
logreg.fit(X_train, y_train)
```

```python
# Generate predictions
y_hat_train = logreg.predict(X_train)
y_hat_test = logreg.predict(X_test)
```

```python

# We could subtract the two columns. If values or equal, difference will be zero. Then count number of zeros.
residuals = np.abs(y_train - y_hat_train)
print(pd.Series(residuals).value_counts())
print('------------------------------------')
print(pd.Series(residuals).value_counts(normalize=True))
# 194 correct, ~ 85% accuracy
```

```python

# We could subtract the two columns. If values or equal, difference will be zero. Then count number of zeros.
residuals = np.abs(y_test - y_hat_test)
print(pd.Series(residuals).value_counts())
print('------------------------------------')
print(pd.Series(residuals).value_counts(normalize=True))
# 62 correct, ~ 82% accuracy
```
